```python
from transformers import AutoTokenizer

# Chargement d'un tokenizer pré-entraîné
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

texts = [
    "source address destination address.",
    "http2 path requester nf type."
]

for text in texts:
    print("\n=== TEXT ===")
    print(text)

    # Tokenisation
    tokens = tokenizer.tokenize(text)
    print("Tokens:", tokens)

    # Conversion en IDs
    ids = tokenizer.convert_tokens_to_ids(tokens)
    print("Token IDs:", ids)

    # (optionnel) version complète avec special tokens
    encoded = tokenizer.encode(text, add_special_tokens=True)
    print("Full encoding (with special tokens):", encoded)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Modèle BERT
MODEL_NAME = "bert-base-uncased"

# Chargement tokenizer + modèle
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModel.from_pretrained(MODEL_NAME)

# Textes à tester
texts = [
    "source address destination address",
    "http2 path requester nf type"
]

# Désactive les gradients (inférence uniquement)
with torch.no_grad():

    for text in texts:

        print(f"\n==============================")
        print(f"TEXT: {text}")

        # Tokenisation
        inputs = tokenizer(text, return_tensors="pt")

        # Affichage tokens
        tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
        print("Tokens :", tokens)

        # Passage dans BERT
        outputs = model(**inputs)

        # Embeddings des tokens
        token_embeddings = outputs.last_hidden_state[0]

        print("\nToken embeddings shape:")
        print(token_embeddings.shape)

        # Affichage embedding de chaque token
        for token, emb in zip(tokens, token_embeddings):

            print(f"\nTOKEN: {token}")
            print("Embedding (first 10 values):")
            print(emb[:10])

        # Embedding global de la séquence ([CLS])
        cls_embedding = token_embeddings[0]

        print("\n[CLS] embedding (first 10 values):")
        print(cls_embedding[:10])
```
